### Defining plot color

```python
x = range(1,10)
y = [ 1,2,5,2,3,8,3,3,5]
plt.bar(x,y)
```

```python
wide = [0.3,.4,.4,.5,.6,.1,.5,.5,.5]
color = ['orange','red','blue','green','pink','skyblue','violet','cyan']   #multiple color it cacn take
plt.bar(x,y, width=wide, color=color, align='center')
```

```python
x1 = range(1,10)
y1 = [3,2,5,2,0.4,2,1,9,5]

plt.plot(x,y, color='red')
plt.plot(x1,y1,color='green')
#plt.bar(x,y,color=['orange']) #multiple plot in a graph 
```
