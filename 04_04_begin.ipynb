### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,4,2,3,7,4,5,3,2]
plt.bar(x,y, color='cyan')
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
z= y.copy()
z=z[0:5]
label_pie= ['raj','rohan','shyam','mohan','neeta']
plt.pie(z, labels=label_pie)

plt.show()
```
