```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] =( 5,4)
```

### Defining axes, ticks, and grids

```python
x = range(1,10)
y = [1,2,4,3,2,0,9,3,1]

fig = plt.figure()
ax = fig.add_axes([5,.1,1,1])   #explore it 
ax.plot(x,y)  #can take multiple args
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([0,9])
ax.set_ylim([0,10])

ax.set_xticks(np.arange(10))
ax.set_yticks(y)

ax.grid()   # check without this code also
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)

```
